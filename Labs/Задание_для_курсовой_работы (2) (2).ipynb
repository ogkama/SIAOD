–ù–µ—Ç —Å–ø–æ—Å–æ–±–∞ –ø–æ–ª—É—á–∏—Ç—å ¬´aaaa¬ª –ø—Ä–∏ –ø–æ–ø—ã—Ç–∫–µ –Ω–∞–ø–µ—á–∞—Ç–∞—Ç—å ¬´aaa¬ª.

–ß—Ç–æ–±—ã –ø–æ–ª—É—á–∏—Ç—å ¬´ababa¬ª –ø—Ä–∏ –ø–æ–ø—ã—Ç–∫–µ –Ω–∞–ø–µ—á–∞—Ç–∞—Ç—å ¬´aababa¬ª, –º–æ–∂–Ω–æ –Ω–∞–∂–∞—Ç—å Backspace –≤–º–µ—Å—Ç–æ –ø–µ—á–∞—Ç–∏ –ø–µ—Ä–≤–æ–≥–æ —Å–∏–º–≤–æ–ª–∞, –∞ –∑–∞—Ç–µ–º –Ω–∞–ø–µ—á–∞—Ç–∞—Ç—å –≤—Å–µ –æ—Å—Ç–∞–≤—à–∏–µ—Å—è —Å–∏–º–≤–æ–ª—ã.

```python
def recursive(s,t,z=0,i=0,option = 1):
    if z < len(s):
        if s[z] == t[i] and option != 3:
            if i == len(t) - 1:
                return True
            else:
                if recursive(s,t,z,i+1,3):
                    return True
        if recursive(s,t,z+3,i,1):
            return True
        if option != 2:
            if recursive(s,t,z+1,i,2):
                return True
    return False  
    
q = int(input())
if (1 <= q <= 3*100000):
    slist = [None for i in range(q)]
    tlist = [None for i in range(q)]
    tr = False
    for i in range(q):
        slist[i] = input()
        if not slist[i]:
            tr = True
            break
        tlist[i] = input()
        if not tlist[i]:
            tr = True
            break
    if tr:
        print("–ù–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π –≤–≤–æ–¥!")
    else:
        for i in range(q):
            if recursive(slist[i],tlist[i]):
                print("YES")
            else:
                print("NO")
else:
    print("–ù–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π –≤–≤–æ–¥!")
```

–í —Ç—Ä–µ—Ç—å–µ–º –ø—Ä–∏–º–µ—Ä–µ –Ω–∞ —Ä—ã–Ω–∫–µ –µ—Å—Ç—å —Ç–æ–ª—å–∫–æ –æ–¥–∏–Ω —Å—É–≤–µ–Ω–∏—Ä —Å—Ç–æ–∏–º–æ—Å—Ç—å—é 8 —Ñ—É–Ω—Ç–æ–≤, –°–∞—Ö–∏—Ä –Ω–µ –º–æ–∂–µ—Ç –µ–≥–æ –∫—É–ø–∏—Ç—å.



```python
def minCost(n,s,a):
    a.sort()   
    t = 0
    newT = 0
    arifm = 0
    k = 1
    for i in range(n):
        newT = a[i] + arifm + i + k*k
        t += newT
        if t > s:
            t = t - newT
            break
        arifm += i
        k += 1   
    print(k - 1, t)
    
n, s = map(int, input().split()) 
if (1 <= n <= 100000) and (1 <= s <= 1000000000):
    a = list(map(int, input().split()))
    if len(a) == n:
        tr = False
        for elem in a:
            if not (1 <= elem <= 100000):
                tr = True
                break
        if tr:
            print("–ù–µ–∫–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π –≤–≤–æ–¥!")
        else:
            minCost(n,s,a)
    else:
        print("–ù–µ–∫–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π –≤–≤–æ–¥!")
else:
    print("–ù–µ–∫–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π –≤–≤–æ–¥!")
```

```python
def perestanovka(n,q):
    x = q[0]
    onPlace = set() #—ç–ª–µ–º–µ–Ω—Ç—ã –Ω–∞ —Å–≤–æ–µ–º –º–µ—Å—Ç–µ
    onPlace.add(x)
    indexToPlace = [] #–∏–Ω–¥–µ–∫—Å—ã, –¥–ª—è —Ä–∞—Å—Å—Ç–∞–Ω–æ–≤–∫–∏ –Ω–æ–≤—ã—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤
    onPlaceIndex = [0] #–∏–Ω–¥–µ–∫—Å—ã —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –Ω–∞ —Å–≤–æ–µ–º –º–µ—Å—Ç–µ
    for i in range(1,n):
        if q[i] > x:
            onPlaceIndex.append(i)
            onPlace.add(q[i])
            x = q[i]
        else:
            indexToPlace.append(i)       
            
    if not indexToPlace: #–µ—Å–ª–∏ —Ä–∞—Å—Å—Ç–∞–Ω–æ–≤–∫–∞ –Ω–µ —Ç—Ä–µ–±—É–µ—Ç—Å—è
        return q,q
    
    options = set() 
    for  i in range(1,x+1):
        options.add(i)
    options = options ^ onPlace
    options = list(options)
    options.sort() #—ç–ª–µ–º–µ–Ω—Ç—ã –¥–ª—è –Ω–æ–≤–æ–π —Ä–∞—Å—Å—Ç–∞–Ω–æ–≤–∫–∏
    
    minPerestanovka = [None for i in range(n)]
    maxPerestanovka = [None for i in range(n)]
    for i in range(n):
        minPerestanovka[i] = q[i]
        maxPerestanovka[i] = q[i]
    
    j = 0
    z = 0
    for i in range(1,n):
        if i == indexToPlace[j]:
            minPerestanovka[i] = options[z] 
            j+=1
            z+=1
        if j == len(indexToPlace):
            break 
    if len(onPlaceIndex) == 1:
        z = 0
    else:
        z = 1
    #print(onPlaceIndex)
    for i in range(1,n):
        if i != onPlaceIndex[z]:
            for j in range(len(options)-1,-1,-1):
                a = options[j] < q[onPlaceIndex[z]]
                if len(onPlaceIndex) != 1:
                    b = i > onPlaceIndex[z] > onPlaceIndex[z-1]
                    c = options[j] < q[onPlaceIndex[z-1]]
                    if (b or c) and a:
                        maxPerestanovka[i] = options[j]
                        options[j] = x
                        break
                elif a:
                    maxPerestanovka[i] = options[j]
                    options[j] = x
                    break
        elif z != len(onPlaceIndex) - 1:
            z += 1
    return minPerestanovka, maxPerestanovka

n = int(input())
if n >= 1:
    q = [None for i in range(n)]
    k = [None for i in range(n)]
    tr = False
    for i in range(n):
        k[i] = int(input())
        if k[i] <= 0:
            tr = True
            break
        q[i] = list(map(int, input().split()))
    if tr:
        print("–ù–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π –≤–≤–æ–¥!")
    else:
        for i in range(n):
            p = perestanovka(k[i],q[i])
            for elem in p[0]:
                print(elem, end = " ")
            print()
            for elem in p[1]:
                print(elem, end = " ")
            print()
else:
    print("–ù–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π –≤–≤–æ–¥!")
```

```python
def story(s,n):
    symb = [[0 for i in range(5)] for j in range(n)] 
    for i in range(n):
        x = s[i].count("a")
        symb[i][0] = x - (len(s[i]) - x) 
        x = s[i].count("b")
        symb[i][1] = x - (len(s[i]) - x) 
        x = s[i].count("c")
        symb[i][2] = x - (len(s[i]) - x) 
        x = s[i].count("d")
        symb[i][3] = x - (len(s[i]) - x)        
        x = s[i].count("e")
        symb[i][4] = x - (len(s[i]) - x)        
    a = [0 for i in range(n)]
    b = [0 for i in range(n)]
    c = [0 for i in range(n)]
    d = [0 for i in range(n)]
    e = [0 for i in range(n)]
    for i in range(n):
        a[i] = symb[i][0]
        b[i] = symb[i][1]
        c[i] = symb[i][2]
        d[i] = symb[i][3]
        e[i] = symb[i][4]  
    indexa = sorted(range(n),key=a.__getitem__)
    indexb = sorted(range(n),key=b.__getitem__)
    indexc = sorted(range(n),key=c.__getitem__)
    indexd = sorted(range(n),key=d.__getitem__)
    indexe = sorted(range(n),key=e.__getitem__)
    indexa.reverse()
    indexb.reverse()
    indexc.reverse()
    indexd.reverse()
    indexe.reverse()
    sm = a[indexa[0]]
    ja = 1
    while sm > 0 and ja < n:  
        sm += a[indexa[ja]]
        ja += 1
    if sm <= 0:
        ja -= 1
    sm = b[indexb[0]]
    jb = 1
    while sm > 0 and jb < n:
        sm += b[indexb[jb]]
        jb += 1
    if sm <= 0:
        jb -= 1
    sm = c[indexc[0]]
    jc = 1
    while sm > 0 and jc < n:   
        sm += c[indexc[jc]] 
        jc += 1
    if sm <= 0:
        jc -= 1
    sm = d[indexd[0]]
    jd = 1
    while sm > 0 and jd < n:      
        sm += d[indexd[jd]]
        jd += 1
    if sm <= 0:
        jd -= 1
    sm = e[indexe[0]]
    je = 1
    while sm > 0 and je < n:  
        sm += e[indexe[je]]
        je += 1
    if sm <= 0:
        je -= 1
    return max(ja,jb,jc,je,jd,0)
    
t = int(input())
if t >= 1:
    n = [None for i in range(t)]
    s = [[] for i in range(t)]
    tr = False
    for i in range(t):
        n[i] = int(input())
        if n[i] <= 0:
            tr = True
            break
        #s = [None for i in range(n[i])]
        for j in range(n[i]):
            s[i].append(input())
    if tr:
        print("–ù–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π –≤–≤–æ–¥!")
    else:
        for i in range(t):
            print(story(s[i],n[i]))
else:
    print("–ù–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π –≤–≤–æ–¥!")
```

```python
def solve(a,b,c):
    n = []
    for s in range(1,81):
        x = b*(s**a) + c
        if x < 0:
            continue
        m = x
        sm = 0
        while m > 0:
            sm += m % 10
            m //= 10
        if sm == s and x < 1000000000:
            n.append(x)
    n.sort()
    return n

a,b,c = map(int, input().split())
if (1 <= a <= 5) and (-1 <= b <= 10000) and (-10000 <= c <= 10000):
    n = solve(a,b,c)
    print(len(n))
    for elem in n:
        print(elem, end=" ")
else:
    print("–ù–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π –≤–≤–æ–¥!")
```

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\

–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
-1

```python
def perestanovka2(n,q):
    s = 0
    minS = 0
    for i in range(n-1):
        s += q[i]
        if s < minS:
            minS = s
    p = [None for i in range(n)]
    p[0]= 1 - minS
    for i in range(1,n):
        p[i] = p[i-1] + q[i-1]
    for i in range(n):
        if p[i] < 1 or p[i] > n:
            return [-1]
    m = set(p)
    if len(m) == n:
        return p
    return [-1]
    
n = int(input())
if (2 <= n <= 2*100000):
    q = list(map(int, input().split()))
    tr = False
    for elem in q:
        if elem <= -n or elem >= n:
            tr = True
            break
    if tr:
        print("–ù–µ–∫–æ—Ä—Ä–µ—Ç–∫–Ω—ã–π –≤–≤–æ–¥!")
    else:
        p = perestanovka2(n,q)
        for elem in p:
            print(elem, end = " ")
else:
    print("–ù–µ–∫–æ—Ä—Ä–µ—Ç–∫–Ω—ã–π –≤–≤–æ–¥!")
```

```python
def wonderful(a,n):
    a = [0] + a
    val = [0 for i in range(n+1)]
    ind = [0 for i in range(n+1)]
    k = [-1 for i in range(n+1)]
    for i in range(1,n+1):
        val[a[i]] = max(val[a[i]],i - ind[a[i]])
        ind[a[i]] = i
        
    for i in range(1,n+1):
        val[i] = max(val[i], n - ind[i] + 1)
        for j in range(val[i],n+1):
            if k[j] != -1:
                continue
            k[j] = i
            
    return k[1:]

t = int(input())
if 1 <= t <= 1000:
    n = [None for i in range(t)]
    a = [None for i in range(t)]
    tr = False
    for i in range(t):
        n[i] = int(input())
        if n[i] <= 0 or n[i] > 3 * 100000:
            tr = True
            break
        a[i] = list(map(int, input().split()))
    if tr:
        print("–ù–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π –≤–≤–æ–¥!")
    else:
        for i in range(t):
            print(wonderful(a[i],n[i]))  
else:
    print("–ù–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π –≤–≤–æ–¥!")
```

```python
def binarSearchRight(a,value):
    left = -1 
    right = len(a) 
    while right > left + 1: 
        mid = (left + right) // 2 
        if a[mid] >= value: 
            right = mid
        else: 
            left = mid 
    return right

def competition(a,batles,n,m):
    result = [0 for i in range(n)]
    for i in range(m):
        deleted = []
        index = binarSearchRight(a,batles[i][0]-1)
        while index < len(a):
            if a[index] > batles[i][1] - 1:
                break
            if a[index] != batles[i][2] - 1:
                result[a[index]] = batles[i][2]
                deleted.append(a[index])
            index += 1
        a = set(a)
        for elem in deleted:
            a.discard(elem)
        a = list(a)  
    return result    
 
n,m = map(int, input().split())
if (2 <= n <= 3*100000) and (1 <= m <= 3*100000):
    a = [i for i in range(n)]
    batles = [None for i in range(m)]
    tr = False
    for i in range(m):
        batles[i] = list(map(int, input().split()))
        if not (1 <=batles[i][0]<batles[i][1]<= n) or not (batles[i][0]<=batles[i][2]<=batles[i][1]):
            tr = True
            break
    if tr:
        print("–ù–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π –≤–≤–æ–¥!")
    else:
        print(competition(a,batles,n,m))    
else:
    print("–ù–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π –≤–≤–æ–¥!")
```
